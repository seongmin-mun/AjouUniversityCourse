# 규칙기반 기계번역
- 언어학적 규칙을 사용하여 소스 언어에서 목표 언어로의 변환을 수행하는 방법이다(Weaver, 1955; Hutchins, 1994).

```python
# 간단한 규칙 기반 번역 시스템
# 한국어에서 영어로 번역하는 간단한 규칙 정의

def korean_to_english(sentence):
    # 사전 정의 (어휘 변환)
    lexicon = {
        "나는": "I",
        "사과를": "an apple",
        "먹었다": "ate",
    }

    # 문장 구조에 따른 규칙 변환 (간단한 SOV 구조 변환)
    def translate_word(word):
        return lexicon.get(word, word)  # 사전에 있으면 번역, 없으면 원래 단어 반환

    # 한국어 문장을 띄어쓰기로 분리 후, 각 단어를 영어로 번역
    words = sentence.split()
    translated_words = [translate_word(word) for word in words]

    # 한국어는 SOV 구조, 영어는 SVO 구조이므로 순서 변환 (주어-동사-목적어)
    if len(translated_words) == 3:
        subject, object_, verb = translated_words  # "나는 사과를 먹었다"
        return f"{subject} {verb} {object_}"      # "I ate an apple"

    return " ".join(translated_words)  # 규칙에 맞지 않는 경우 단순히 이어 붙이기

# 테스트
korean_sentence = "나는 사과를 먹었다"
english_translation = korean_to_english(korean_sentence)
print("Translated Sentence:", english_translation)
```
