# 예제기반 기계번역
- 미리 저장된 번역 예제(병렬 코퍼스)를 이용해 비슷한 문장을 찾고, 이를 기반으로 번역하는 방법이다(Makoto, 1984; Sato, 1995).
- 최조 편 집거리 참고논문: https://www.kci.go.kr/kciportal/ci/sereArticleSearch/ciSereArtiView.kci?sereArticleSearchBean.artiId=ART003103346

```python
# 입력 문장과 예제 문장의 거리를 계산하는 함수 / 음절 기반

def edit_dist(str1, str2):
    dp = [[0] * (len(str2)+1) for _ in range(len(str1) + 1)]
    for i in range(1, len(str1)+1):
        dp[i][0] = i
    for j in range(1, len(str2)+1):
        dp[0][j] = j

    for i in range(1, len(str1)+1):
        for j in range(1, len(str2)+1):
            if str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]

            else:
                dp[i][j] = min(dp[i-1][j-1], dp[i-1][j], dp[i][j-1]) + 1

    return dp[-1][-1]
```

```python
# 번역 예제 데이터베이스
example_database = {
    "나는 사과를 먹었다.": "I ate an apple.",
    "나는 바나나를 먹는다.": "I eat a banana.",
    "나는 물을 마신다.": "I drink water."
}


# 가장 유사한 예문을 찾아 번역하는 함수
def translate_using_example(input_sentence):
    # 예제 데이터베이스에서 가장 유사한 문장 찾기
    most_similar_sentence = None
    lowDist = 100
    translation = "No translation found."

    for example_sentence, example_translation in example_database.items():
        editDist = edit_dist(input_sentence, example_sentence)

        if editDist < lowDist:
            lowDist = editDist
            most_similar_sentence = example_sentence
            translation = example_translation

    print(f"Most similar sentence: {most_similar_sentence} (Distance: {lowDist})")
    return translation

# 테스트
korean_sentence = "나는 사과를 먹는다."
english_translation = translate_using_example(korean_sentence)
print("Translated Sentence:", english_translation)
```
