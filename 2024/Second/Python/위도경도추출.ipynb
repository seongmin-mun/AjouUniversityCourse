### 데이터 불러와서 위도 경도 추출

```python
fr = open("./Starbucks_utf8.csv", 'r', encoding='utf-8')
contents = fr.readlines()
fr.close()

addressList = []
for content in contents:
    contentSplit = content.split(",")
    if contentSplit[0].strip() == "address":
        pass
    else:
        eachAddress = contentSplit[0].split("(")[0].strip()
        addressList.append(eachAddress)
        print(eachAddress)
        
latitudeList = []
longitudeList = []
for each in addressList:
    print(each)
    try:  
        geo = geo_local.geocode(each)
        latitudeList.append(geo.latitude)
        longitudeList.append(geo.longitude)
    except:
        latitudeList.append("NA")
        longitudeList.append("NA")


addressDF = pd.DataFrame({'주소': addressList,'위도':latitudeList,'경도':longitudeList})

addressDF.to_csv('addressDF.csv')
```
