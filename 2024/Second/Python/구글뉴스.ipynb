## loop사용해서 페이지 이동

```python
for i in range(4,10):
    #다음 페이지로 넘어가기
    search = browser.find_element(By.XPATH,'//*[@id="botstuff"]/div/div[3]/table/tbody/tr/td['+str(i)+']/a')
    search.click()
    time.sleep(2) # 로딩 시간을 위한 대기 
    
    # 페이지 소스 출력
    html = browser.page_source
    html_source = BeautifulSoup(html, 'html.parser')
    print(html_source)
    
    #뉴스 타이틀 가져오기
    newsTitle = html_source.find_all("div", class_ = "n0jPhd ynAwRc MBeuO nDgy9d")
    for eachTitle in newsTitle:
        titleList.append(eachTitle.get_text())
        print(eachTitle.get_text())
        
    #뉴스 요약 가져오기
    newsSummary = html_source.find_all("div", class_ = "GI74Re nDgy9d")
    for eachSummary in newsSummary:
        summaryList.append(eachSummary.get_text())
        print(eachSummary.get_text())
```

```python
#가져온 정보 확인 가져오기
print(len(titleList))
print(len(summaryList))
```

```python
googleNewsDF = pd.DataFrame({'title': titleList,'summary':summaryList})
googleNewsDF.to_csv('./Data/googleNewsDF.csv')
```
