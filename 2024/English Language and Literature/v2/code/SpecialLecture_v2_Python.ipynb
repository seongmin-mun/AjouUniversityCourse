## 선행 연구 논문들의 결과
- Slater and Feinman (1985): 여성의 이름이 음절 수와 분절음의 수가 더 많으며 공명음(/m/, /n/, /l/, /r/, /ŋ/, /y/, /w/)으로 끝나는 경우가 더 많다.
- Barry and Harper (1995): 여성의 이름은 공명음으로, 남성의 이름은 조음과정에서 조음기관에 의해 소리의 흐름이 방해를 받는 소리인 저해음으로 끝나는 경우가 많았다.
- Moorthy et al. (2018): 이름이 /a/로 끝나면 여성의 이름, 공명음, 파열음(i.e., 소리의 흐름이 조음기관에 의해 완전히 닫혔다가 열리면서 나오는 소리), 마찰음(i.e., 파열음처럼 완전히 닫히지 않고 조음기관이 닫히기 직전까지 가까워지며 그 과정에 소음을 내는 소리)으로 끝나면 남성의 이름으로 예측 가능하다는 것을 보고 하였다.

### 모음의 분류 기준
- ㄱ. 혀의 높낮이에 따른 분류 (high, low)
- ㄴ. 혀의 전후에 따른 분류 (front, back)

![Consonants](https://seongmin-mun.github.io/AjouUniversityCourse/2024/English%20Language%20and%20Literature/v2/image/Consonants.png)

### 조음 위치에 따른 자음의 구분
- ㄱ. 양순음 (Bilabial): 윗입술과 아랫입술이 닿으면서 나는 소리
- ㄴ. 순치음 (Labiodental): 윗니가 아랫입술에 살짝 닿으면서 나는 소리
- ㄷ. 치간음 (Interdental): 윗니와 아랫니 사이에 혀끝이 내밀어지거나 접근되어서 나는 소리
- ㄹ. 치경음 (Alveolar): 혀 끝이 치경에 접근되면서 나는 소리
- ㅁ. 경구개음 (Palatal): 혀 앞부분이 경구개에 접근되면서 나는 소리
- ㅂ. 연구개음 (Velar): 혀 뒷부분이 연구개에 접근되면서 나는 소리
- ㅅ. 성문음 (Glottal): 공기 흐름이 성문을 통과하면서 마찰이 생길 때 나는 소리

### 조음 방법에 따른 자음의 구분
- ㄱ. 파열음 (Plosive): 공기 흐름 차단 방식이 소리를 생성하는 데 가장 중요한 역할을 담당. 소리를 잠시 멈추게 한 후 터뜨리듯 공기를 내뱉는 소리
- ㄴ. 마찰음 (Fricative): 조음 기관들이 서로 근접하면서 공기 흐름을 방해하여 나오는 마찰이 소리를 생성하는 과정
- ㄷ. 파찰음 (Affricate): 구강 내에서 차단되었던 공기를 조금씩 통과시키는 방식으로 만들어지는 소리
- ㄹ. 비음 (Nasal): 공기의 흐름이 비강으로 흘러들어가면서 생성되는 소리
- ㅁ. 유음 (Liquid): 공기 흐름에 방해가 있기는 하지만, 마찰을 일으킬 정도는 아닌 소리. /l/은 모음처럼 음절을 관장하는 자질을 가진다.
- ㅂ. 활음 (Glide): 혀의 조음 위치가 미끄러지듯 이동하는 변화로 인하여 생기는 소리

```python
import re

fileDir = "drive/My Drive/v2/data/names_ny.txt"

fr = open(fileDir, 'r')
contents = fr.readlines()
fr.close()

BoyNames = []
GirlNames = []

for content in contents:
  contentSplit = content.replace("\n","").split(",")
  Gender = contentSplit[1]
  Year = int(contentSplit[2])
  Name = contentSplit[3].lower()
  Freq = int(contentSplit[4])
  if 1995 < Year and Year < 2000:
    for i in range(0,Freq):
      if Gender == "M":
        BoyNames.append(Name)
      else:
        GirlNames.append(Name)

print(len(BoyNames))
print(len(GirlNames))
```

### CMU 사전 불러오기
- CMUdict (the Carnegie Mellon Pronouncing Dictionary): https://github.com/cmusphinx/cmudict

```python
import re

fileDir = "drive/My Drive/v2/data/cmudict.dict"

fr = open(fileDir, 'r')
contents = fr.readlines()
fr.close()

CMUdict = dict()

for content in contents:
  contentSplit = content.replace("\n","").split(" ")
  value = ""
  if " # " in content:
    value = content.replace("\n","").replace(contentSplit[0]+" ","").split(" # ")[0]
  else:
    value = content.replace("\n","").replace(contentSplit[0]+" ","")

  value = re.sub("\d+","",value)

  CMUdict[contentSplit[0]] = value
```

```python
num = 0
for key, value in CMUdict.items():
  if num < 200:
    print(key,"  ",value)
  num += 1
```

### 무작위 표본 추출을 통해 성비에 따른 이름 데이터 사이즈 균등화하기

```python
import random

def RandomSelect(dataList, size):
  dataSize = size

  dataListRD = []
  for i in range(0, dataSize):
      randomNum = random.randrange(0, len(dataList))
      dataListRD.append(dataList[randomNum])
      del dataList[randomNum]

  return dataListRD
```

```python
BoyNames_10000 = RandomSelect(BoyNamesRe, 10000)
GirlNames_10000 = RandomSelect(GirlNamesRe, 10000)

print(len(BoyNames_10000), BoyNames_10000[1:10])
print(len(GirlNames_10000), GirlNames_10000[1:10])
```

### 기저형(written) -> 표면형(spoken)

```python
def G2Peng(dataList, CMUdict, IPAdict):

  outcomeList = []

  for input in dataList:

    outcome = ""

    words = input.split(" ")
    for word in words:
      output = ""
      if CMUdict.get(word) != None:
        if " " in CMUdict.get(word):
          CMUdictSplit = CMUdict.get(word).split(" ")
          for eachCMU in CMUdictSplit:
            if IPAdict.get(eachCMU) != None:
              output = output + " " + IPAdict.get(eachCMU)
            else:
              output = output + " " + eachCMU
        else:
          if IPAdict.get(CMUdict.get(word)) != None:
            output = output + " " + IPAdict.get(CMUdict.get(word))
          else:
            output = output + " " + CMUdict.get(word)
      else:
        output = " OOV"
      outcome = outcome + output.strip() + " "
    outcomeList.append(outcome.strip())

  return outcomeList
```

```python
BoyNamesSpoken = G2Peng(BoyNames_10000, CMUdict, IPAdict)
GirlNamesSpoken = G2Peng(GirlNames_10000, CMUdict, IPAdict)

print(len(BoyNamesSpoken), BoyNamesSpoken[1:10])
print(len(GirlNamesSpoken), GirlNamesSpoken[1:10])
```

### 자음 관련 변수 생성하기
- 공명음의 비율
- 장애음의 비율
- 공명음으로 끝나는지
- 장애음으로 끝나는지

```python
#자음 관련 정보 기준들
sonorantConsonants = ["m","n","l","r","w","j","ŋ"]
obstruentConsonants = ["p","b","f","v","ɵ","ð","t","d","s","z","ʃ","ʒ","ʧ","ʤ","k","g","h"]
```

```python
#sonorantConsonantsRate
SCR_boy = []
SCR_girl = []
#obstruentConsonantsRate
OCR_boy = []
OCR_girl = []
#sonorantConsonantsEnd
SCE_boy = []
SCE_girl = []
#obstruentConsonantsEnd
OCE_boy = []
OCE_girl = []


for i in range(0,len(BoyNamesSpoken)):
  BoyNamesSpokenSplit = BoyNamesSpoken[i].split(" ")
  #sonorantConsonantsRate
  #obstruentConsonantsRate
  SC = 0
  OC = 0
  for each in BoyNamesSpokenSplit:
    if each in sonorantConsonants:
      SC += 1
    elif each in obstruentConsonants:
      OC += 1
  SCR_boy.append(SC/BoyNamesSegments[i])
  OCR_boy.append(OC/BoyNamesSegments[i])

  #sonorantConsonantsEnd
  if BoyNamesSpokenSplit[-1] in sonorantConsonants:
    SCE_boy.append(1)
  else:
    SCE_boy.append(0)

  #obstruentConsonantsEnd
  if BoyNamesSpokenSplit[-1] in obstruentConsonants:
    OCE_boy.append(1)
  else:
    OCE_boy.append(0)




for i in range(0,len(GirlNamesSpoken)):
  GirlNamesSpokenSplit = GirlNamesSpoken[i].split(" ")
  #sonorantConsonantsRate
  #obstruentConsonantsRate
  SC = 0
  OC = 0
  for each in GirlNamesSpokenSplit:
    if each in sonorantConsonants:
      SC += 1
    elif each in obstruentConsonants:
      OC += 1
  SCR_girl.append(SC/GirlNamesSegments[i])
  OCR_girl.append(OC/GirlNamesSegments[i])

  #sonorantConsonantsEnd
  if GirlNamesSpokenSplit[-1] in sonorantConsonants:
    SCE_girl.append(1)
  else:
    SCE_girl.append(0)

  #obstruentConsonantsEnd
  if GirlNamesSpokenSplit[-1] in obstruentConsonants:
    OCE_girl.append(1)
  else:
    OCE_girl.append(0)


#sonorantConsonantsRate
print(len(SCR_boy), SCR_boy[1:10])
print(len(SCR_girl), SCR_girl[1:10])
#obstruentConsonantsRate
print(len(OCR_boy), OCR_boy[1:10])
print(len(OCR_girl), OCR_girl[1:10])
#sonorantConsonantsEnd
print(len(SCE_boy), SCE_boy[1:10])
print(len(SCE_girl), SCE_girl[1:10])
#obstruentConsonantsEnd
print(len(OCE_boy), OCE_boy[1:10])
print(len(OCE_girl), OCE_girl[1:10])
```

### 모음 관련 변수 생성하기
- 전설 모음의 비율
- 중설 모음의 비율
- 후설 모음의 비율
- 모음으로 끝나는지

```python
frontVowels = ["i","ɪ","e","ɜ","æ"]
centralVowels = ["ə"]
backVowels = ["u","ʊ","o","ɔ","ɒ"]
vowels = ["i","ɪ","e","ɜ","æ","ə","u","ʊ","o","ɔ","ɒ"]
```

```python
#frontVowelsRate
FVR_boy = []
FVR_girl = []
#centralVowelsRate
CVR_boy = []
CVR_girl = []
#backVowelsRate
BVR_boy = []
BVR_girl = []
#vowelsEnd
VE_boy = []
VE_girl = []


for i in range(0,len(BoyNamesSpoken)):
  BoyNamesSpokenSplit = BoyNamesSpoken[i].split(" ")
  #frontVowelsRate
  #centralVowelsRate
  #backVowelsRate
  FV = 0
  CV = 0
  BV = 0
  for each in BoyNamesSpokenSplit:
    if each in frontVowels:
      FV += 1
    elif each in centralVowels:
      CV += 1
    elif each in backVowels:
      BV += 1
  FVR_boy.append(FV/BoyNamesSegments[i])
  CVR_boy.append(CV/BoyNamesSegments[i])
  BVR_boy.append(BV/BoyNamesSegments[i])

  #vowelsEnd
  if BoyNamesSpokenSplit[-1] in vowels:
    VE_boy.append(1)
  else:
    VE_boy.append(0)




for i in range(0,len(GirlNamesSpoken)):
  GirlNamesSpokenSplit = GirlNamesSpoken[i].split(" ")
  #frontVowelsRate
  #centralVowelsRate
  #backVowelsRate
  FV = 0
  CV = 0
  BV = 0
  for each in GirlNamesSpokenSplit:
    if each in frontVowels:
      FV += 1
    elif each in centralVowels:
      CV += 1
    elif each in backVowels:
      BV += 1
  FVR_girl.append(FV/GirlNamesSegments[i])
  CVR_girl.append(CV/GirlNamesSegments[i])
  BVR_girl.append(BV/GirlNamesSegments[i])

  #vowelsEnd
  if GirlNamesSpokenSplit[-1] in vowels:
    VE_girl.append(1)
  else:
    VE_girl.append(0)


#frontVowelsRate
print(len(FVR_boy), FVR_boy[1:10])
print(len(FVR_girl), FVR_girl[1:10])
#centralVowelsRate
print(len(CVR_boy), CVR_boy[1:10])
print(len(CVR_girl), CVR_girl[1:10])
#backVowelsRate
print(len(BVR_boy), BVR_boy[1:10])
print(len(BVR_girl), BVR_girl[1:10])
#vowelsEnd
print(len(VE_boy), VE_boy[1:10])
print(len(VE_girl), VE_girl[1:10])
```
