## 자료형

### 숫자형
- 숫자 형태로 이루어진 자료형

### 문자형
- 문자, 단어 등으로 구성된 문자들의 집합

### 리스트 자료형
- 1, 3, 5, 7, 9 숫자 모음

### 집합 자료형
- 집합에 관련된 것을 쉽게 처리하기 위해 만든 자료형

집합 자료형의 특징
- 중복을 허용하지 않는다.
- 순서가 없다(Unordered).

### 딕셔너리 자료형
- 대응 관계를 나타낼 수 있는 자료형

### 불(bool) 자료형
- 참(True)과 거짓(False)을 나타내는 자료형
