## 자료형

### 숫자형
- 숫자 형태로 이루어진 자료형

```python

a = 456
a = -173
a = 0

a = 1.2
a = -3.45
a = 4.24E3   
a = 4.24e-3

a = 3
b = 5

print(a + b)
b - a
a * b
a / b

a = 3
b = 5
print(a ** b)

7 % 2
print(2 % 7)

print(7 // 2)
7 // 2
```

```python
# 정수형: 정수를 뜻하는 자료형

a = 456
a = -173
a = 0

# 실수형: 소수점이 포함된 숫자

a = 1.2
a = -3.45
a = 4.24E3   
a = 4.24e-3

# 기본 연산
# 사칙연산(+, -, *, /)

a = 3
b = 5

print(a + b)
b - a
a * b
a / b

# x의 y제곱을 나타내는 ** 연산자

a = 3
b = 5
print(a ** b)

# 나눗셈 후 나머지를 반환하는 % 연산자

7 % 2
print(2 % 7)

# 나눗셈 후 몫을 반환하는 // 연산자

print(7 // 2)
7 // 2
```

### 문자형
- 문자, 단어 등으로 구성된 문자들의 집합

### 리스트 자료형
- 1, 3, 5, 7, 9 숫자 모음

### 집합 자료형
- 집합에 관련된 것을 쉽게 처리하기 위해 만든 자료형

집합 자료형의 특징
- 중복을 허용하지 않는다.
- 순서가 없다(Unordered).

### 딕셔너리 자료형
- 대응 관계를 나타낼 수 있는 자료형

### 불(bool) 자료형
- 참(True)과 거짓(False)을 나타내는 자료형

## 제어문

### for문
- 반복해서 작업을 수행해야 할 경우 사용

### if문
- 주어진 조건을 판단한 후 그 상황에 맞게 처리해야 할 경우 사용
